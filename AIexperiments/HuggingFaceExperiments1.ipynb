```python
#Tryin Zero-shot classification which is to classify texts that haven't been labelled.
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "Man I love to spark up in the morning.",
    candidate_labels=["cannabis", "tea", "coffee"],
)
```

```python
classifier(
    "I want to become a great AI Engineer!",
    candidate_labels=["ambition", "fear", "lust", "greed"]
)
```

```python
classifier(
    "I desire to escape the simulation.",
    candidate_labels=["intellect", "emotion",  "madness"]
)
```

```python
#Now we will try Mask filling which means to fill in blanks in a given text.
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

```python
unmasker = pipeline("fill-mask", model='bert-base-cased')
```

```python
unmasker("The third planet in the solar system is [MASK].", top_k=2)
```

```python
#Named Entity Recognition is a task where the model has to find which parts of the input text correspond to entities such as person, locations, or organizations.
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Isaac and I live in Alberta, Canada.")
```

```python
#Question Answering

from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Isaac and I currently am unemployed.",
)
```
